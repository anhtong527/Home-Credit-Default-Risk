## 2.1. bureau balance ==> bureau ==> df

### bureau.csv

* All client's previous credits provided by other financial institutions that were reported to Credit Bureau (for clients who have a loan in our sample).
* For every loan in our sample, there are **as many rows as number of credits the client had in Credit Bureau before the application date**.
* It is matched by SK_ID_CURR with application datasets. The relationship between train/test and  bureau is 1:M, each client can have 1 or multiple loans previously.

### bureau_balance.csv
* Monthly balances of previous credits in Credit Bureau.
* This table has **one row for each month of history of every previous credit** reported to Credit Bureau – i.e the table has (#loans in sample * # of relative previous credits * # of months where we have some history observable for the previous credits) rows.
* It related to bureau via SK_ID_BUREAU.

### approach:
calculate mean / median / max of each statistical columns out of these both df --> features. But it only works with numerical features, eg. client with SK_ID_CURR 666 had 7 credits in bureau df, and when grouped by 1 ID, we will not be able to show a CREDIT_ACTIVE column (Closed / Active for different previous credits).

But we can calculate the total number of previous credits taken by each client.

```python
n_previous_loans = bureau.groupby('SK_ID_CURR')['SK_ID_BUREAU'].count().reset_index().rename(columns={'SK_ID_BUREAU': 'NUM_PREVIOUS_LOANS'})
n_previous_loans.head()
```

```python
## add NUM_PREVIOUS_LOANS column to df
df = df.merge(n_previous_loans, how='left', on='SK_ID_CURR')
df.head()
```

### 2.1.1 bureau_balance ==> bureau

```python
bureau_balance.head()
```

```python
bur_bal_grb = bureau_balance.groupby('SK_ID_BUREAU').mean().add_prefix('BUR_BAL_').reset_index()
bur_bal_grb.head()
```

```python
bureau = bureau.merge(bur_bal_grb, how='left', on='SK_ID_BUREAU')
bureau.drop('SK_ID_BUREAU', inplace=True, axis=1)
bureau.head()
```

```python
print(bureau.shape)
bureau['SK_ID_CURR'].nunique()
### --> need to groupby to make SK_ID_CURR unique
```

### 2.1.2. bureau ==> df

```python
bur_grb = bureau.groupby('SK_ID_CURR').mean().add_prefix('BUR_').reset_index()
bur_grb.head()
```

```python
print(bur_grb.shape)
bur_grb['SK_ID_CURR'].nunique()
```

```python
print(df.shape)
df = df.merge(bur_grb, how='left', on='SK_ID_CURR')
print(df.shape)
df.head()
```

### 2.2.1 POS_CASH_balance ==> previous_application
**POS_CASH_balance.csv**
* Monthly balance snapshots of previous POS (point of sales) and cash loans that the applicant had with Home Credit.
* This table has **one row for each month of history of every previous credit** in Home Credit (consumer credit and cash loans) related to loans in our sample – i.e. the table has (#loans in sample * # of relative previous credits * # of months in which we have some history observable for the previous credits) rows.

```python
pos_cash_balance.head()
```

```python
pos_cash_balance.drop('SK_ID_CURR', axis=1, inplace=True)
```

```python
pos_cash_grb = pos_cash_balance.groupby('SK_ID_PREV').mean().add_prefix('POS_').reset_index()
pos_cash_grb.head()
```

```python
prev_app = prev_app.merge(pos_cash_grb, how='left', on='SK_ID_PREV')
prev_app.head()
```

### 2.2.2 installments_payments ==> previous_application
**installments_payments.csv**
* Repayment history for the previously disbursed credits in Home Credit related to the loans in our sample.
* There is a) **one row for every payment** that was made plus b) **one row each for missed payment**.
* One row is equivalent to one payment of one installment OR one installment corresponding to one payment of one previous Home Credit credit related to loans in our sample.

```python
ins_pay.head()
```

```python
ins_pay.drop('SK_ID_CURR', axis=1, inplace=True)
```

```python
ins_pay_grb = ins_pay.groupby('SK_ID_PREV').mean().add_prefix('INS_PAY_').reset_index()
ins_pay_grb.head()
```

```python
prev_app = prev_app.merge(ins_pay_grb, how='left', on='SK_ID_PREV')
prev_app.head()
```

### 2.2.3 credit_card_balance ==> previous_application
**credit_card_balance.csv**
* Monthly balance snapshots of previous credit cards that the applicant has with Home Credit.
* This table has **one row for each month of history of every previous credit** in Home Credit (consumer credit and cash loans) related to loans in our sample – i.e. the table has (#loans in sample * # of relative previous credit cards * # of months where we have some history observable for the previous credit card) rows.

```python
credit_card_balance.head()
```

```python
credit_card_balance.drop('SK_ID_CURR', axis=1, inplace=True)
```

```python
credit_card_grb = credit_card_balance.groupby('SK_ID_PREV').mean().add_prefix('CRED_CARD_').reset_index()
credit_card_grb.head()
```

```python
prev_app = prev_app.merge(credit_card_grb, how='left', on='SK_ID_PREV')
prev_app.head()
```
